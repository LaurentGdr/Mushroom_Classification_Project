# Mushroom Edibility Classification: Problem Introduction

The task involves using data about mushrooms to determine whether they are **edible** or **poisonous** based on their physical and ecological characteristics. This is a classification problem where we aim to predict the **class** of a mushroom (edible or poisonous) using a set of features, such as:

- **Cap characteristics**: Size, shape, color, surface texture.
- **Gill features**: Spacing, attachment, color.
- **Stem attributes**: Height, width, surface texture, and color.
- Other factors like the presence of a ring, spore print color, habitat, and season.

The goal is to build a **predictive model** using **multiple classification algorithms** to identify the most effective one for the task. Examples of algorithms we will explore include:

- **Random Forest**: Robust for handling complex feature relationships.
- **Logistic Regression**: A baseline model for binary classification.
- **Decision Trees**: Simple and interpretable.
- **K-Nearest Neighbors (KNN)**: Non-parametric and instance-based learning.

---

## Why Is This Problem Important?

This type of classification has real-world implications, particularly for **foraging safety** and **environmental studies**:

1. **Safety**: Identifying poisonous mushrooms is critical to avoid health risks.
2. **Automation**: Automating the identification process reduces reliance on manual expertise.

The dataset consists of features that can be categorical or numerical. Proper preprocessing and feature engineering are essential before feeding data into machine learning models. 

We will focus on:
- Understanding the data structure.
- Preparing the data for machine learning (cleaning and encoding).
- Training and validating models using multiple algorithms.
- Evaluating and comparing performance across algorithms to identify the best one.


# Feature Value Key

## Cap Features:
- **Cap Shape:**
  $$ \text{bell} = b, \quad \text{conical} = c, \quad \text{convex} = x, \quad \text{flat} = f, \quad \text{knobbed} = k, \quad \text{sunken} = s $$
- **Cap Surface:**
  $$ \text{fibrous} = f, \quad \text{grooves} = g, \quad \text{scaly} = y, \quad \text{smooth} = s $$
- **Cap Color:**
  $$ \text{brown} = n, \quad \text{buff} = b, \quad \text{cinnamon} = c, \quad \text{gray} = g, \quad \text{green} = r, \quad \text{pink} = p, $$
  $$ \text{purple} = u, \quad \text{red} = e, \quad \text{white} = w, \quad \text{yellow} = y $$

---

## Bruises and Odor:
- **Bruises?:**
  $$ \text{bruises} = 1, \quad \text{no} = 0 $$
- **Odor:**
  $$ \text{almond} = a, \quad \text{anise} = l, \quad \text{creosote} = c, \quad \text{fishy} = y, \quad \text{foul} = f, $$
  $$ \text{musty} = m, \quad \text{none} = n, \quad \text{pungent} = p, \quad \text{spicy} = s $$

---

## Gill Features:
- **Gill Attachment:**
  $$ \text{attached} = 1, \quad \text{free} = 0 $$
- **Gill Spacing:**
  $$ \text{close} = 0, \quad \text{crowded} = 1 $$
- **Gill Size:**
  $$ \text{broad} = 0, \quad \text{narrow} = 1 $$
- **Gill Color:**
  $$ \text{black} = k, \quad \text{brown} = n, \quad \text{buff} = b, \quad \text{chocolate} = h, \quad \text{gray} = g, \quad \text{green} = r, $$
  $$ \text{orange} = o, \quad \text{pink} = p, \quad \text{purple} = u, \quad \text{red} = e, \quad \text{white} = w, \quad \text{yellow} = y $$

---

## Stalk Features:
- **Stalk Shape:**
  $$ \text{enlarging} = 0, \quad \text{tapering} = 1 $$
- **Stalk Root:**
  $$ \text{bulbous} = b, \quad \text{club} = c, \quad \text{cup} = u, \quad \text{equal} = e, \quad \text{rhizomorphs} = z, $$
  $$ \text{rooted} = r, \quad \text{missing} = ? $$

  $$ \text{fibrous} = f, \quad \text{scaly} = y, \quad \text{silky} = k, \quad \text{smooth} = s $$

  $$ \text{fibrous} = f, \quad \text{scaly} = y, \quad \text{silky} = k, \quad \text{smooth} = s $$

  $$ \text{brown} = n, \quad \text{buff} = b, \quad \text{cinnamon} = c, \quad \text{gray} = g, \quad \text{orange} = o, \quad \text{pink} = p, $$

  $$ \text{brown} = n, \quad \text{buff} = b, \quad \text{cinnamon} = c, \quad \text{gray} = g, \quad \text{orange} = o, \quad \text{pink} = p, $$
  $$ \text{red} = e, \quad \text{white} = w, \quad \text{yellow} = y $$

---

## Veil and Ring Features:
- **Veil Color:**
  $$ \text{brown} = n, \quad \text{orange} = o, \quad \text{white} = w, \quad \text{yellow} = y $$
- **Ring Number:**
  $$ \text{none} = n, \quad \text{one} = o, \quad \text{two} = t $$
- **Ring Type:**
  $$ \text{cobwebby} = c, \quad \text{evanescent} = e, \quad \text{flaring} = f, \quad \text{large} = l, \quad \text{none} = n, $$
  $$ \text{pendant} = p, \quad \text{sheathing} = s, \quad \text{zone} = z $$

---

## Spore, Population, and Habitat:

  $$ \text{black} = k, \quad \text{brown} = n, \quad \text{buff} = b, \quad \text{chocolate} = h, \quad \text{green} = r, \quad \text{orange} = o, $$
  $$ \text{purple} = u, \quad \text{white} = w, \quad \text{yellow} = y $$

  $$ \text{abundant} = a, \quad \text{clustered} = c, \quad \text{numerous} = n, \quad \text{scattered} = s, \quad \text{several} = v, \quad \text{solitary} = y $$

## Performance Metrics: Accuracy and F1 Score

## Accuracy

$$
\text{Accuracy} = \frac{\text{Number of Correct Predictions}}{\text{Total Number of Predictions}}
$$

Accuracy measures how many predictions the model got right out of all predictions made. While it's a simple metric, it may not be reliable when dealing with **imbalanced datasets** (e.g., when one class is much more frequent than another).

---

## F1 Score

$$
\text{F1 Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

The F1 score is the **harmonic mean** of Precision and Recall. It balances the trade-off between the two and is particularly useful when the dataset is imbalanced.

- **Precision**: The proportion of correctly predicted positive cases out of all predicted positive cases.
  $$
  \text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}
  $$
- **Recall**: The proportion of correctly predicted positive cases out of all actual positive cases.
  $$
  \text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}
  $$

### Cross-Validation

```python
from sklearn.model_selection import cross_val_score

models = {
    'Random Forest': rf,
    'Decision Tree': dt,
    'Logistic Regression': lr,
    'KNN': knn
}


results = {}
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='f1')  # F1 score
    results[name] = scores


for name, scores in results.items():
    print(f"{name}: Mean Accuracy = {scores.mean():.3f}, Standard Deviation = {scores.std():.3f}")

best_model_name = max(results, key=lambda k: results[k].mean())
best_model = models[best_model_name]

print(f"\nBest Model: {best_model_name}")
```

### ROC

```python
from sklearn.metrics import roc_curve, roc_auc_score, RocCurveDisplay
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt


lr_probs = lr.predict_proba(X_test)[:, 1]  
rf_probs = rf.predict_proba(X_test)[:, 1]  
dt_probs = dt.predict_proba(X_test)[:, 1]  
knn_probs = knn.predict_proba(X_test)[:, 1]  


lr_auc = roc_auc_score(y_test, lr_probs)
rf_auc = roc_auc_score(y_test, rf_probs)
dt_auc = roc_auc_score(y_test, dt_probs)
knn_auc = roc_auc_score(y_test, knn_probs)


lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_probs)
dt_fpr, dt_tpr, _ = roc_curve(y_test, dt_probs)
knn_fpr, knn_tpr, _ = roc_curve(y_test, knn_probs)

plt.figure(figsize=(10, 6))
plt.plot(lr_fpr, lr_tpr, label=f'Logistic Regression (AUC = {lr_auc:.2f})', color='blue')
plt.plot(rf_fpr, rf_tpr, label=f'Random Forest (AUC = {rf_auc:.2f})', color='orange')
plt.plot(dt_fpr, dt_tpr, label=f'Decision Tree (AUC = {dt_auc:.2f})', color='green')
plt.plot(knn_fpr, knn_tpr, label=f'KNN (AUC = {knn_auc:.2f})', color='red')

 
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```

The ROC curve shows near-perfect performance for Random Forest, Decision Tree, and KNN with an AUC of 1.00, indicating that these models classify the mushrooms without any false positives or negatives.


Logistic Regression also performs exceptionally well, with an AUC of 0.99, reflecting minimal classification errors.
